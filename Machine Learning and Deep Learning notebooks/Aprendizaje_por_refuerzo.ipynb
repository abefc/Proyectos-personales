##1.Definición

El aprendizaje por refuerzo (RL) es una rama del aprendizaje automático en la que un agente aprende a tomar decisiones interactuando con su entorno. A diferencia del aprendizaje supervisado (que usa datos etiquetados) o no supervisado, en RL el agente no recibe ejemplos de la respuesta “correcta”, sino que aprende por ensayo y error recibiendo señales de recompensa. En otras palabras, se trata de un enfoque centrado en la toma de decisiones secuenciales bajo incertidumbre, donde el objetivo es maximizar una recompensa acumulada a lo largo del tiempo.

El aprendizaje por refuerzo se usa en problemas como control de robots, juegos o sistemas de recomendación, donde el agente debe descubrir cuál es la mejor estrategia (política) para lograr una meta a largo plazo.

Los dos elementos principales a destacar son:


*   Agente: Es la entidad (un algoritmo o sistema autónomo) que toma decisiones y actúa en el entorno. Por ejemplo, un robot, un software de juego o un vehículo autónomo pueden ser agentes.
*   Entorno: Es el “mundo” o espacio de problemas donde opera el agente, incluyendo las reglas y variables que definen cómo cambian los estados y recompensas. El entorno responde a las acciones del agente proporcionando una nueva observación (estado) y una recompensa.

Estos dos componentes representan prácticamente en lo que consiste el aprendizaje por refuerzo, una interacción contínua del agente con el entorno que le permite explorar cuáles serán las mejores decisiones en el futuro a base de fallos. Este 'núcleo' del aprendizaje por refuerzose repite en el tiempo de la siguiente manera:

1.   El agente observa el estado actual del entorno (su situación presente).
1.   Según su política, el agente elige una acción a ejecutar en ese estado.
2.   El entorno transiciona al siguiente estado como consecuencia de esa acción y genera una recompensa (valor numérico) que refleja el “éxito” o “fracaso” de la acción.
1.   El agente recibe la recompensa y actualiza su política o sus estimaciones internas; luego repite el ciclo desde el nuevo estado.

Cualquier algoritmo de RL se basa en esta dinámica. Por ejemplo IBM describe:“El entorno proporciona información sobre su estado actual. El agente emplea esa información para determinar qué acción emprender. Si esa acción obtiene una señal de recompensa, el agente se ve animado a volver a realizarla en estados futuros similares”. Con el tiempo, repitiendo el proceso, el agente va mejorando su comportamiento buscando maximizar la recompensa total.









##2.Estado, acción y recompensa

Estos tres términos tienen una gran relevancia cuando hablamos de aprendizaje pro refuerzo, ya que sientan las bases para la dinámica del ciclo que antes hemos mencionado. Son los siguientes:

*   Estado: es la descripción completa de la situación actual del entorno, tal como la percibe el agente. Es la información que el agente observa en cada paso de tiempo (por ejemplo, la posición de un robot, la posición de las piezas en un juego, la temperatura de una habitación, etc.). Dentro del estado hay que mencionar el espacio de estados, que es el conjunto de todos los estados posibles que el entorno puede adoptar. En un entorno discreto, esto corresponde a todas las configuraciones factibles. Por ejemplo, en un laberinto el espacio de estados serían todas las casillas posibles; en un problema más complejo, podría ser una combinación de múltiples variables. En la práctica, el estado es la información que usa el agente para tomar acción.
*   Acción: es cualquier decisión o movimiento que el agente puede realizar en un estado dado. Por ejemplo, un robot puede tener acciones como avanzar, girar a la derecha o saltar; en un videojuego, podría escoger entre moverse, disparar, saltar, etc. Las acciones son las herramientas con las que el agente interactúa y cambia el estado del entorno. Dentro de la acción hay que describir el espacio de acciones, que es el conjunto de todas las acciones posibles que el agente puede elegir en cada estado. Puede variar según el estado (algunas acciones pueden no estar disponibles en ciertos estados).
*   La recompensa es una señal numérica que el entorno otorga al agente tras cada acción, indicando lo “bueno” o “malo” de dicha acción en ese momento. Puede ser positiva (premio), negativa (castigo) o cero. Por ejemplo, en un juego el agente puede recibir +1 por cada objetivo alcanzado o -1 por cada error. Estas recompensas guían el aprendizaje: el agente aprenderá a favorecer las acciones que históricamente llevaron a recompensas altas y evitar las que dieron recompensas bajas. El objetivo final del agente es maximizar la suma de estas recompensas a largo plazo (la recompensa acumulada o retorno). En otras palabras, no se trata de optimizar la recompensa inmediata, sino la ganancia total futura. Un agente óptimo es aquel cuya estrategia consigue el mayor retorno esperado posible.

Con todos estos aspectos teóricos, podemos acercarnos más a como y el por qué de las acciones del agente.


##3.Política y experiencia

Estos dos puntos, aunque no tan relacionados, vienen directamente afectados por lo que hemos definido en el anterior apartado. Los tratamos por separado.

La política (se suele nombrar con π) de un agente es su estrategia o regla de decisión, que indica qué acción elegir en cada estado. En la teoría, la política mapea estados a acciones. Como dice IBM: “la política define el comportamiento del agente, asignando estados ambientales percibidos a acciones específicas que el agente debe realizar”. Las políticas pueden clasificarse en dos tipos:

*   Política determinista: define una acción fija para cada estado. Es decir, dado un estado s, la política determinista π(s) retorna una sola acción. No hay aleatoriedad en la selección.
*   Política estocástica (no deterministas): asigna una distribución de probabilidad sobre las acciones posibles en cada estado. Bajo esta política, una misma observación puede llevar a diferentes acciones en distintas ocasiones, de acuerdo a ciertas probabilidades. Por ejemplo, una política estocástica podría indicar que en el estado A el agente elija la acción X con 70% de probabilidad y la acción Y con 30%. Estas políticas estocásticas son útiles porque fomentan la exploración del entorno (probar diversas acciones) y evitan quedarse atrapado en soluciones subóptimas. En cambio, las políticas deterministas son más sencillas y directas cuando el entorno es predecible.

En cuanto a la experiencia, cada vez que el agente ejecuta una acción en el entorno y observa la respuesta, genera una experiencia. Concretamente, una experiencia típica se puede representar como la 4-tupla (s, a, r, s′): el estado anterior s, la acción tomada a, la recompensa recibida r, y el nuevo estado s′. En una trayectoria completa, el agente acumula una secuencia de estas tuplas. Según Google Developers, una trayectoria (secuencia de transiciones) está formada por tuplas donde “cada tupla corresponde al estado, la acción, la recompensa y el estado siguiente”. El agente usa estas experiencias (a veces almacenándolas) para ajustar su política y mejorar gradualmente.

Sabiendo esto, podemos adentrarnos a explicar como actua un agente de comienzo a fin.


##4.Episodio y trayectoria

Los términos que vamos a explicar, como hemos dicho antes, abarcan la actuación del agente de comienzo a fin, a más y menos detalle están:

* Episodio: es una ejecución completa desde un estado inicial hasta que se
alcanza un estado final. En problemas episódicos, existe un final definido (por ejemplo, finalizar un juego con victoria o derrota). Un episodio corresponde a todos los pasos de tiempo que dura dicha ejecución. En el ejemplo del robot, cada recorrido del inicio hasta llegar a la meta sería un episodio.
* Trayectoria: es la secuencia de experiencias ocurridas durante un episodio. Es decir, el conjunto de tuplas (s, a, r, s′) a lo largo de un episodio. En la práctica se usa a veces como sinónimo de episodio, pero en general un episodio es el evento completo y la trayectoria es la ruta de experiencias dentro de él.

Por tanto, episodio subraya el comienzo y fin de una ejecución, mientras que trayectoria enfatiza la serie de pasos intermedios. En aprendizaje por refuerzo, el agente suele repetir muchos episodios (reiniciando el entorno) para aprender de cada trayectoria experimentada.

##5.Estructura subyacente de un entorno

En teoría, el entorno de RL se modela como un Proceso de Decisión de Markov (MDP). Un MDP es un modelo matemático para problemas de decisión secuenciales en pasos de tiempo discretos, donde las transiciones de estado cumplen la propiedad de Markov, esto significa que el siguiente estado depende solo del estado actual y la acción tomada, no de pasos pasados. Google lo describe como “un modelo de toma de decisiones en el que se toman acciones para navegar por una secuencia de estados, asumiendo la propiedad de Markov, y estas transiciones devuelven una recompensa”.

 En un MDP se definen explícitamente que es el conjunto de estados, el conjunto de acciones, la función de transición (qué estado sigue a cada par estado-acción) y la función de recompensa.

Un ejemplo muy sencillo pdría ser el siguiente
->imaginamos un entorno con dos estados S={A,B} y dos acciones posibles {IrA,
IrB}. Las reglas (modelo) podrían ser -> desde el estado A, la acción IrB lleva al estado B con recompensa +1, mientras que IrA deja al agente en A con recompensa 0. Desde el estado B, la acción IrA lleva a A con recompensa +2 y IrB lo mantiene en B con recompensa –1.

Este mini-MDP define claramente cómo cambia el estado y qué recompensa se recibe por cada acción. La estructura del MDP garantiza que el agente pueda predecir las consecuencias de sus acciones (según las reglas) y planificar hacia objetivos lejanos. Casi todos los algoritmos de RL asumen esta estructura subyacente.

##6.Objetivo del agente

El objetivo principal del agente en RL es aprender una política óptima que maximice el retorno o recompensa acumulada a lo largo del tiempo. En lugar de medir precisión de predicción (como en supervisado), el agente busca la secuencia de acciones que produzca la mayor ganancia total. Dicho de otro modo, el agente “quiere maximizar la recompensa acumulativa” del entorno. Esto implica considerar las recompensas futuras, no solo la inmediata. Por ejemplo, si un agente puede obtener +5 hoy o +10 mañana, una buena política lo llevará a sacrificar recompensas pequeñas inmediatas por las mayores a largo plazo, siempre que el retorno esperado sea mayor.

Dentro de esta definición hemos mencionado el retorno:
* Retorno: es la suma total de recompensas que el agente obtiene desde un cierto punto en adelante. Generalmente se representa como la suma acumulada de recompensas futuras. Formalmente, si el agente recibe recompensas:
$$
\
R = r_0 + \gamma r_1 + \gamma^2 r_2 + \dots + \gamma^N r_N
\
$$
,donde γ es el factor de descuento. Este cálculo aparece en definiciones formales de RL; por ejemplo, Google indica que el retorno es “la suma de todas las recompensas que el agente espera recibir cuando sigue la política desde el estado hasta el final del episodio”, y lo ejemplifica con la fórmula anterior
. El retorno refleja entonces el rendimiento total de una trayectoria de acciones bajo la política del agente. El agente intenta maximizar el valor esperado de este retorno al ajustar su política.

Como vemos, nos ha surgido otro término a explicar dentro de esta definición, que es el factor de descuento:
* Factor de descuento: El factor de descuento γ (un número entre 0 y 1) establece cuánto valora el agente las recompensas lejanas en el futuro. Cada recompensa obtenida k pasos adelante se multiplica por γ^k. Esto tiene dos propósitos, uno estabilizar el cálculo del retorno en problemas continuos (evitando sumas infinitas) y dos reflejar la preferencia por recompensas más inmediatas. Por ejemplo, con 𝛾=0.9, una recompensa futura de 10 unids. a 2 pasos recibe un “peso” de 0.9^2=0.81, contribuyendo 8.1 al retorno en vez de 10. Si γ es cercano a 1, el agente considera casi igual las recompensas lejanas y cercanas (es decir, planifica a largo plazo). Si γ es más bajo (por ejemplo 0.5), el agente prioriza mucho más las recompensas inmediatas. Como muestra la fórmula del retorno, γ controla la influencia de cada recompensa futura. En resumen, el factor de descuento balancea entre explorar beneficios futuros y explotar recompensas inmediatas.

Con todo esto abarcado, queda por analizar un último punto en el que veremos los valores de un estado y de estado-acción.

##7.Valor estado y estado-acción

Podemos diferir en dos valores relevantes en este caso:

* La función de valor de estado V(s) mide la calidad de un estado s. Es el valor esperado del retorno que obtendría el agente estando en s y siguiendo su política actual a partir de ese punto. Explicado de una forma más simple, V(s) indica “qué tan bueno” es el estado s considerando las recompensas futuras probabilísticas.  Es decir, V(s) integra las recompensas inmediatas y las esperadas en el futuro desde ese estado. El agente busca estados con alto V(s) ya que conducen a más recompensa esperada más adelante.

* La función Q o valor estado-acción Q(s,a) extiende la idea anterior considerando una acción concreta. Es el valor esperado del retorno si el agente toma la acción a en el estado s y luego sigue la política. Más simple con otras palabras, Q(s,a) predice la recompensa total esperada al comenzar en s, ejecutar a y luego actuar según la política. El valor Q(s,a) es muy útil porque permite comparar directamente qué acción es mejor en cada estado. Una política óptima suele seleccionar la acción con mayor Q(s,a) en cada estado. Así, Q(s,a) guía al agente sobre qué acción tomar considerando las recompensas inmediatas y futuras.

3. la posibilidad de crear envoltorios personalizados (wrappers) y vectores de entornos.

En definitiva, Gymnasium es el marco base para desarrollar y evaluar agentes de RL.

En cuanto al entorno CliffWalking-v0 es un gridworld (se puede pensar como una tabla) discreto de 4×12 casillas adaptado del libro de Sutton y Barto. El agente comienza en la casilla inferior izquierda (S, coordenadas [3,0], estado codificado 36) y debe llegar a la casilla inferior derecha (G, coordenadas [3,11], estado 47). Entre S y G hay un “precipicio” que ocupa las casillas [3,1..10] (fila inferior, columnas 1 a 10). Si el agente se mueve a una casilla de precipicio, recibe una fuerte penalización y es reiniciado al inicio.

* Espacio de acciones: 4 acciones discretas posibles (tipo Discrete(4)): 0=arriba, 1=derecha, 2=abajo, 3=izquierda. El agente puede moverse en esas cuatro direcciones en cada paso.
* Espacio de estados: 48 estados (Discrete(48)) que codifican la posición del agente en la cuadrícula. Se asigna un entero de la forma row×12+col al estado; por ejemplo la casilla inicial [3,0] es 36 (3×12+0). En realidad, hay 3×12+1=37 posiciones utilizables: el agente no puede permanecer en los estados de precipicio ni en la meta (este último hace terminar el episodio), de modo que quedan 37 estados alcanzables.
* Recompensa: En cada paso válido el agente recibe recompensa −1. Si el agente se mueve a una casilla de precipicio, recibe recompensa −100 y es automáticamente devuelto a la casilla inicial. Así, evitar el precipicio es esencial para maximizar la recompensa a largo plazo (que sería más cercana a cero).
* Terminación del episodio: El episodio termina cuando el agente alcanza la meta (estado 47). En ese momento se señala terminated=True. De forma implícita, también se podría truncar el episodio si se impone un límite de pasos (usando TimeLimit), pero en el funcionamiento por defecto solo el llegar a G pone fin al episodio.

Para concluír con esto, CliffWalking es un entorno determinista (salvo que se active is_slippery, deshabilitado por defecto) con un espacio de estados y acciones discretos pequeños, recompensa negativa en cada paso y un gran castigo por caer. Es un caso clásico para comparar algoritmos off-policy (Q-learning) vs on-policy (SARSA)

Ahora cabe mencionar dos métodos fundamentales que permiten la interacción con cualquier entorno de Gymnasium:

* reset(): Inicializa o reinicia el entorno para comenzar un nuevo episodio. No suele recibir parámetros (aunque se puede pasar seed para reproducibilidad). Devuelve dos valores: la observación inicial (observation) y un diccionario info opcional. En CliffWalking, por ejemplo, reset() coloca al agente en el estado inicial [36] (casilla S). El diccionario info puede contener información adicional (por defecto vacío en este entorno). En Gymnasium, se suele llamar observation, info = env.reset(); cada vez que termina un episodio (al llegar a meta o truncar), debe usarse reset() para comenzar otro episodio. Por ejemplo, DataCamp señala: “.reset() restablece el entorno a un estado inicial. Por defecto, el entorno se inicializa con un estado aleatorio. Puedes usar un parámetro seed con .reset() para iniciar siempre en el mismo estado”.
* step(action): Avanza el entorno un paso dinámico según la acción del agente. Se le pasa como parámetro la acción seleccionada (action), y devuelve cinco valores: la próxima observación (estado) observation, la recompensa obtenida reward, dos flags booleanos terminated y truncated, y el diccionario info.
Más concretamente:
  * observation: estado resultante tras aplicar la acción (un entero en CliffWalking).
  * reward: recompensa obtenida tras la acción (en CliffWalking será −1 o −100 según el caso)
  * terminated: indicador de que se alcanzó un estado terminal del entorno (aquí, llegar a la meta); si es True, el episodio terminó correctamente.
  * truncated: indicador de terminación por límite externo (por ejemplo un límite de pasos). Normalmente es False salvo que se use un wrapper TimeLimit o condiciones de corte; indica un final artificial fuera del MDP original
  * info: diccionario con información adicional (usado para diagnósticos, suele incluir "TimeLimit.truncated" en Gym antiguo, pero en Gymnasium ya no es necesario).

Por ejemplo, en cada iteración típica se hace algo como:

```
action = env.action_space.sample()  #acción que hace el agente
observation, reward, terminated, truncated, info = env.step(action)
```

Según la documentación, step ejecuta la transición y luego “retorna el siguiente estado, la recompensa, si se llegó a un estado terminal (terminated) o si fue truncado, y info”. Si terminated o truncated son True, se debe llamar a env.reset() para iniciar un nuevo episodio.

Dicho de forma básica, reset() prepara el entorno y da el primer estado, mientras que step(action) aplica la acción, devuelve el nuevo estado, la recompensa, y señala si el episodio terminó o fue truncado.

##2.Políticas manuales con TimeLimit

Para probar “a mano” sin aprender, vamos a envolver nuestro entorno en un límite de pasos y definir dos funciones que actúen como políticas. Vamos a crear dos entornos distintos, esto con el objetivo de reducir el tiempo de ejecución, ya que si solo usasemos un entorno con render_mode en los 500 pasos con 100 episodios sería eterna la ejecución. Creamos un entorno con render y otro sin render para poder medir rápido:

````python
import gymnasium as gym
from gymnasium.wrappers import TimeLimit

#creamos el entorno con límite de 500 pasos por episodio y render
base_env = gym.make('CliffWalking-v0')
env = TimeLimit(
    gym.make('CliffWalking-v0', render_mode='human'),
    max_episode_steps=500
)
# Entorno SIN render, solo para medir rápido
eval_env = TimeLimit(
    gym.make('CliffWalking-v0'),
    max_episode_steps=500
)
````

Con esto, podemos pasar a concretar cada una de las políticas

###2.1 Política aleatoria

La política aleatoria no mira el estado. Simplemente pide una acción al espacio de acciones:

````python
def random_policy(env, state):
    """
    devuelve siempre una acción al azar
    del espacio de acciones.
    """
    return env.action_space.sample()
````

###2.2 Política óptima manual

Como conocemos la cuadrícula, escribimos una función que, paso a paso, sigue la ruta segura (subir al borde, avanzar y bajar al final):

````python
def optimal_policy(env, state):
    """
    Sube a fila 2, avanza hasta columna 11 y baja a la meta.
    """
    ncols = 12
    row, col = divmod(state, ncols)

    # 1) En (3,0) subir
    if (row, col) == (3, 0):
        return 0          # arriba

    # 2) En fila 2 avanzar hasta col 11 (no 10)
    if row == 2 and col < 11:
        return 1          # derecha

    # 3) En (2,11) bajar a la meta
    if (row, col) == (2, 11):
        return 2          # abajo

    # 4) Cualquier otro caso (p. ej. ya en meta): mover derecha
    return 1
````

###2.3 Evaluación sobre 100 episodios

Definimos una función genérica para correr una política durante N episodios y calcular la recompensa media:

````python
def evaluate_policy(env, policy_fn, episodes=100, render_one=False):
    rewards = []
    for ep in range(episodes):
        state, _ = env.reset()
        done = False
        total_r = 0

        while not done:
            action = policy_fn(env, state)
            state, r, terminated, truncated, _ = env.step(action)
            total_r += r
            done = terminated or truncated
            if render_one and ep == 0:
                env.render()

        rewards.append(total_r)

    avg = sum(rewards) / len(rewards)
    std = (sum((r-avg)**2 for r in rewards)/len(rewards))**0.5
    return avg, std
````

Y finalmente lo ejecutamos (tarda un ratillo)

````python
avg_rand, std_rand = evaluate_policy(eval_env, random_policy,
                                     episodes=100, render_one=False)
avg_opt,  std_opt  = evaluate_policy(eval_env, optimal_policy,
                                     episodes=100, render_one=False)

print(f"Aleatoria → {avg_rand:.2f} ± {std_rand:.2f}")
print(f"Óptima    → {avg_opt:.2f}  ± {std_opt:.2f}")
````

Los números hablan por sí solos. Con la política aleatoria el agente acumula, de media, unos -4 800 puntos por episodio (desviación ≈ 1 400).

¿Por qué tanto castigo? Durante los 500 pasos que permite el wrapper TimeLimit el agente deambula sin un criterio, pisa el precipicio en muchas ocasiones (-100 cada vez) y, aun cuando no cae, sigue recibiendo -1 por cada movimiento. El resultado típico es la suma de 500 penalizaciones pequeñas (-500) más decenas de caídas, de ahí la media tan negativa y la gran variabilidad, en unos episodios se despeña más que en otros.

La política “óptima” hard-coded, en cambio, finaliza cada partida con -13 puntos exactos y desviación nula. Sube un paso, avanza junto al borde superior hasta la columna 11 y baja a la meta, son trece movimientos seguros, sin precipicio y sin truncamiento. Con ese simple conocimiento del entorno el rendimiento mejora tres órdenes de magnitud respecto a la estrategia ciega.

En resumen, CliffWalking muestra cómo la falta de planificación combina una penalización constante (-1 por paso) con castigos puntuales muy altos (-100) y hunde el retorno; basta una ruta informada para eliminar casi todo el daño. El experimento ilustra la importancia, en aprendizaje por refuerzo, de aprender o incorporar rápidamente una política que evite transiciones de alto riesgo en entornos con recompensas tan asimétricas.

##3.Algoritmos clásicos

Antes de la irrupción del Deep RL, la investigación se centraba en métodos tabulares que mantienen una tabla de valores para cada estado (o par estado-acción). Los más citados son:
* Programming Dynamics: Son métodos que encuentran políticas óptimas basándose en un modelo completo del entorno. Operan mediante la iteración de la evaluación y mejora de políticas (Policy Iteration) o resolviendo directamente la ecuación de Bellman óptima para los valores de los estados (Value Iteration). Destacan:
  * Policy Iteration: Alterna «evaluar» la política actual y «mejorarla» hasta converger.
  * Value Iteration: Actualiza directamente la ecuación de Bellman óptima sobre V(s).

* TD (Temporal-Difference): Son algoritmos de aprendizaje que estiman valores directamente a partir de la experiencia, sin requerir un modelo del entorno. Actualizan las estimaciones de valor basándose en otras estimaciones aprendidas. Destacan:
  * SARSA: Algoritmo on-policy, aprende con la acción real que ejecuta.
  * Q-Learning: Algoritmo off-policy, actualiza con max a ′Q(s ′,a ′).
  * Expected SARSA, Double Q, etc. : 	Variaciones para reducir sesgo o varianza.

En nuestro caso, elegimos Q-Learning porque es:
1. sencillo de implementar
2. converge a la política óptima en entornos finitos
3. sirve de referencia para comparar con métodos profundos.

###3.1 Q-Learning

Como hemos mecionado brevemente, el Q-Learning es un algoritmo fundamental de aprendizaje por refuerzo off-policy. Su objetivo es encontrar la política óptima, es decir, las mejores acciones a tomar en cada estado para maximizar la recompensa total a largo plazo, independientemente de la política que se esté siguiendo en un momento dado.

1. Inicialización de la Tabla Q:
Al principio, creamos una tabla Q donde se almacenan los valores Q. Estos valores representan la recompensa esperada al ejecutar una acción específica desde un estado determinado.
Q(s,a)←0∀s∈S,a∈A Inicializamos todos los valores Q a cero para cada combinación de estado (s) y acción (a). S es el conjunto de estados y A el conjunto de acciones.

2. Bucle de Episodios:
El entrenamiento se estructura en episodios. Cada episodio es una secuencia de interacciones entre el agente y el entorno que finaliza al alcanzar un estado terminal o un límite de tiempo.
Reiniciar al estado inicial s: Al comienzo de cada episodio, el agente se posiciona en un estado inicial (predefinido o aleatorio).

3. Bucle de Pasos (dentro de cada episodio):
Dentro de cada episodio, el agente interactúa con el entorno paso a paso:
Selección de acción a (Política ϵ-greedy):
Para decidir qué acción tomar en el estado actual s, el agente usa una política ϵ-greedy. Esta estrategia equilibra la exploración (probar acciones nuevas) y la explotación (tomar la acción que, hasta ahora, parece la mejor).

###3.2 Implementación

El algoritmo lo programaremos íntegramente a pelo, partiendo de una tabla Q(s,a) inicializada a cero y actualizándola paso a paso con la regla de Q-Learning. Vamos a recurrir a una política ϵ-greedy (explora con probabilidad ϵ y, en otro caso, elige la acción con mayor Q) para garantizar que cada par estado-acción se visite suficientes veces y así satisfacer las hipótesis de convergencia.

Los hiperparámetros los fijaremos de forma conservadora, α=0,1 para que los valores se ajusten con suavidad, γ=0,99 porque los episodios duran a lo sumo 500 pasos y conviene valorar casi por igual recompensa presente y futura, y un ϵ que arranca en 1 y decae geométricamente hasta 0,01 para ir pasando gradualmente de la exploración a la explotación.

Lo aplicamos sobre CliffWalking-v0, entorno utilizado en el apartado anterior, un escenario clásico para medir la diferencia entre comportamientos on-policy y off-policy. El límite de 500 pasos impuesto con el wrapper TimeLimit convierte cada partida en un episodio acotado y aporta una referencia clara, incluso sin despeñarse, una política que deambule pierde ya 500 puntos, así que cualquier mejora es fácil de cuantificar, y así también podemos evaluarlo respecto a las políticas anteriores.

````python
import random
import numpy as np

#hiperparametros
ALPHA      = 0.1
GAMMA      = 0.99
EPSILON    = 1.0
EPS_DECAY  = 0.9995
MIN_EPS    = 0.01

env = TimeLimit(gym.make("CliffWalking-v0"), max_episode_steps=500)
nS, nA = env.observation_space.n, env.action_space.n
Q = np.zeros((nS, nA))          #tabla Q(s,a)

def epsilon_greedy(state):
    if random.random() < EPSILON:
        return env.action_space.sample()
    return np.argmax(Q[state])

#entrenamiento
EPISODES = 5_000
for _ in range(EPISODES):
    s, _ = env.reset()
    done = False
    while not done:
        a  = epsilon_greedy(s)
        s2, r, term, trunc, _ = env.step(a)
        td_target = r + GAMMA * np.max(Q[s2])
        Q[s, a]  += ALPHA * (td_target - Q[s, a])
        s = s2
        done = term or trunc
        # update ε
        global EPSILON
        EPSILON = max(MIN_EPS, EPSILON * EPS_DECAY)

def run_policy(episodes=100):
    successes = 0
    returns   = []
    for _ in range(episodes):
        s,_ = env.reset()
        done = False
        G = 0
        while not done:
            a = np.argmax(Q[s])      # política greedy aprendida
            s, r, term, trunc, _ = env.step(a)
            G += r
            done = term or trunc
        returns.append(G)
        if term: successes += 1
    return successes, np.mean(returns)

succ, avg_ret = run_policy()
print(f"Resuelto {succ}/100 episodios · recompensa media = {avg_ret:.1f}")
````

Tras 5 000 episodios de entrenamiento la tabla Q converge y, al ejecutar la política codiciosa resultante en 100 ensayos, el agente llega a la meta en los 100 episodios y obtiene siempre una recompensa de –13. Ese valor coincide exactamente con el coste teórico de la ruta segura (trece movimientos × -1) y la desviación típica es nula porque la política aprendida es determinista y nunca pisa el precipicio.

El resultado obtenido es el mismo que en el algoritmo óptimo del apartado anterior, un algoritmo diseñado expclusivamente para este problema teniendo en cuenta todos los factores.

El experimento nos muestra, en definitiva, cómo un algoritmo clásico y muy compacto como Q-Learning basta para resolver completamente un MDP discreto cuando el espacio de estados es manejable y la política dispone de la exploración necesaria para descubrir la estrategia libre de riesgo, igualando el resultado d eun algoritmo diseñado únicamente para este fin.

### 1. Algoritmos de Deep Reinforcement Learning

El aprendizaje por refuerzo profundo (Deep Reinforcement Learning, DRL) combina el marco del aprendizaje por refuerzo tradicional con redes neuronales profundas, lo que permite a los agentes aprender directamente a partir de entradas de alta dimensión, como imágenes o estados complejos. A diferencia del aprendizaje supervisado, donde se necesita un conjunto de datos etiquetado, DRL aprende por ensayo y error, recibiendo recompensas o penalizaciones según las acciones que ejecuta en un entorno.

Este paradigma se basa en el Proceso de Decisión de Markov (MDP), donde un agente observa un estado del entorno, ejecuta una acción, recibe una recompensa y transita a un nuevo estado. El objetivo es encontrar una política óptima que maximice la recompensa acumulada a lo largo del tiempo, ponderada por un factor de descuento que da más importancia a las recompensas inmediatas.

En DRL existen dos grandes familias de algoritmos:

- **Deep Q-Networks (DQN)**: Son métodos basados en valores. Usan una red neuronal para aproximar la función Q(s, a), que estima la recompensa esperada de realizar una acción en un estado y seguir luego una política óptima. Emplean técnicas como experiencia repetida (replay buffer) y redes objetivo para estabilizar el aprendizaje. Ejemplos destacados son DQN, Double DQN (que reduce el sobreajuste), Dueling DQN (que separa el valor del estado y la ventaja de cada acción) y Prioritized Experience Replay, que mejora la eficiencia del muestreo. Son adecuados para entornos con espacios de acción discretos y definidos.

- **Policy Gradient**: Son métodos que optimizan directamente la política del agente, sin necesidad de estimar explícitamente el valor de cada acción. En lugar de buscar la mejor acción desde la Q-function, la política se representa como una red neuronal que produce distribuciones de probabilidad sobre las acciones. Ejemplos clásicos son REINFORCE, y versiones más estables como Actor-Critic, A2C/A3C, PPO (Proximal Policy Optimization), que restringe los cambios bruscos en la política, o SAC (Soft Actor-Critic), que maximiza simultáneamente la recompensa y la entropía para fomentar la exploración. Estos métodos son más apropiados para espacios de acción continuos o de gran tamaño y permiten representar políticas estocásticas.

Mientras DQN aprende “cuánto vale hacer algo”, Policy Gradient aprende directamente “qué hacer”. Ambos enfoques tienen fundamentos comunes (como la ecuación de Bellman), pero difieren en cómo se modela y mejora la política del agente.

### 3. Descripción del algoritmo

El algoritmo Deep Q-Network (DQN) es una mejora del algoritmo clásico Q-Learning, creado para enfrentar problemas donde el espacio de estados es muy grande o incluso continuo, lo que hace imposible usar una tabla para almacenar los valores de acción-estado Q(s,a). Para esto, DQN utiliza redes neuronales profundas que sirven como aproximadores de la función Q(s,a).

El proceso comienza con un agente que interactúa con un entorno. En cada paso del proceso, el agente observa el estado actual s del entorno y debe elegir una acción a. Esta elección no es arbitraria; el agente sigue una política de exploración y explotación llamada epsilon-greedy, que consiste en elegir la acción con el mejor valor estimado la mayor parte del tiempo (explotación), pero también de vez en cuando probar acciones aleatorias (exploración) para descubrir nuevas estrategias. La probabilidad de exploración está controlada por el parámetro epsilon, que suele empezar alto y disminuir progresivamente.

Después de ejecutar la acción a, el entorno devuelve una recompensa r y un nuevo estado s'. Estos datos forman una experiencia completa (s, a, r, s'), que se guarda en una memoria llamada replay buffer o buffer de experiencia. Guardar estas experiencias permite que, más adelante, el agente pueda entrenar la red neuronal usando datos de diferentes momentos y romper la correlación temporal que existe si solo usara los datos en orden cronológico, lo cual mejora la estabilidad y eficiencia del aprendizaje.

La red neuronal profunda que se utiliza en DQN toma como entrada el estado s y devuelve los valores Q estimados para todas las acciones posibles en ese estado. Es decir, predice qué tan buena es cada acción en ese momento. Sin embargo, para evitar que el entrenamiento sea inestable o divergente, DQN utiliza dos redes neuronales con la misma arquitectura:

- La red principal Q(s,a; θ) con parámetros θ, que es la que se entrena continuamente y genera las predicciones.

- La red objetivo Q(s,a; θ−) con parámetros θ−, que es una copia de la red principal pero cuyos parámetros se mantienen fijos durante un número determinado de pasos. Cada cierto tiempo, los parámetros θ de la red principal se copian a la red objetivo. Esto ayuda a que el cálculo de las metas para el entrenamiento sea más estable.

El objetivo del entrenamiento es minimizar la diferencia entre las predicciones de la red principal y el valor esperado de la recompensa futura, definido por la ecuación de Bellman adaptada a DQN:

Aquí, gamma es un factor de descuento que valora cuánto se priorizan las recompensas futuras respecto a las inmediatas (un valor entre 0 y 1).

Para poder entrenar una red neuronal, necesitamos una función de pérdida o coste (loss or cost function), la cual definimos como el cuadrado de la diferencia entre ambos lados de la ecuación de Bellman:

image.png

El proceso de entrenamiento en cada iteración incluye los siguientes pasos:

1. **Selección de acción**: El agente elige una acción a en el estado actual s siguiendo la política epsilon-greedy basada en los valores que devuelve la red principal.

2. **Interacción con el entorno**: Ejecuta la acción a, recibe la recompensa r y observa el nuevo estado s'.

3. **Almacenamiento de experiencia**: La tupla (s, a, r, s') se guarda en el replay buffer.

4. **Muestreo**: Se selecciona aleatoriamente un conjunto pequeño de experiencias (minibatch) del replay buffer para el entrenamiento.

5. **Cálculo del valor objetivo**: Para cada experiencia del minibatch, se calcula el valor objetivo Q target usando la red objetivo.

6. **Cálculo de la pérdida**: Se calcula la función de pérdida como el error cuadrático medio (MSE) entre los valores Q estimados por la red principal y los valores objetivo.

7. **Actualización de la red principal**: Se ajustan los pesos de la red principal para minimizar la función de pérdida, utilizando técnicas de retropropagación y optimizadores como Adam o SGD.

8. **Actualización de la red objetivo**: Cada cierto número de pasos, se copian los pesos de la red principal a la red objetivo para mantener la estabilidad del aprendizaje.

Además, el parámetro epsilon se va reduciendo gradualmente para que, al principio, el agente explore muchas acciones diferentes, y a medida que aprende, se enfoque más en explotar lo aprendido para maximizar las recompensas.

Gracias a este método, DQN puede aprender a resolver tareas complejas en las que el agente debe tomar decisiones en entornos con muchos estados posibles, como en el famoso problema CartPole. Allí, el agente aprende a mover un carrito para mantener un palo equilibrado, mejorando su desempeño poco a poco hasta lograr mantener el palo estable durante largos periodos.

````python
import os
os.environ['TF_USE_LEGACY_KERAS'] = '1'

import tensorflow as tf
import numpy as np
from tf_agents.agents.dqn import dqn_agent
from tf_agents.networks import q_network
from tf_agents.utils import common
from tf_agents.environments import suite_gym, tf_py_environment
from tf_agents.replay_buffers import tf_uniform_replay_buffer
from tf_agents.policies import random_tf_policy
from tf_agents.trajectories import trajectory
from tf_agents.metrics import tf_metrics
from tf_agents.drivers import dynamic_step_driver
from tf_agents.policies import policy_saver

# ---------- Configuración general ----------
env_name = "CartPole-v1"
num_iterations = 20000
initial_collect_steps = 1000
collect_steps_per_iteration = 1
replay_buffer_max_length = 100000

batch_size = 64
learning_rate = 1e-3
log_interval = 1000
eval_interval = 5000
fc_layer_params = (100, 50)

# ---------- Crear entornos ----------
train_env = tf_py_environment.TFPyEnvironment(suite_gym.load(env_name))
eval_env = tf_py_environment.TFPyEnvironment(suite_gym.load(env_name))

# ---------- Crear QNetwork ----------
q_net = q_network.QNetwork(
    input_tensor_spec=train_env.observation_spec(),
    action_spec=train_env.action_spec(),
    fc_layer_params=fc_layer_params
)

# ---------- Crear agente ----------
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
global_step = tf.compat.v1.train.get_or_create_global_step()

agent = dqn_agent.DqnAgent(
    time_step_spec=train_env.time_step_spec(),
    action_spec=train_env.action_spec(),
    q_network=q_net,
    optimizer=optimizer,
    epsilon_greedy=0.1,
    n_step_update=1,
    target_update_tau=0.05,
    target_update_period=200,
    td_errors_loss_fn=common.element_wise_squared_loss,
    gamma=0.99,
    train_step_counter=global_step
)
agent.initialize()

# ---------- Replay buffer ----------
replay_buffer = tf_uniform_replay_buffer.TFUniformReplayBuffer(
    data_spec=agent.collect_data_spec,
    batch_size=train_env.batch_size,
    max_length=replay_buffer_max_length
)

# ---------- Políticas ----------
random_policy = random_tf_policy.RandomTFPolicy(train_env.time_step_spec(), train_env.action_spec())
eval_policy = agent.policy
collect_policy = agent.collect_policy

# ---------- Recolección inicial ----------
def collect_step(environment, policy, buffer):
    time_step = environment.current_time_step()
    action_step = policy.action(time_step)
    next_time_step = environment.step(action_step.action)
    traj = trajectory.from_transition(time_step, action_step, next_time_step)
    buffer.add_batch(traj)

for _ in range(initial_collect_steps):
    collect_step(train_env, random_policy, replay_buffer)

# ---------- Dataset para entrenamiento ----------
dataset = replay_buffer.as_dataset(
    num_parallel_calls=3,
    sample_batch_size=batch_size,
    num_steps=2
).prefetch(3)
iterator = iter(dataset)

# ---------- Evaluación ----------
def compute_avg_return(environment, policy, num_episodes=10):
    total_return = 0.0
    for _ in range(num_episodes):
        time_step = environment.reset()
        episode_return = 0.0
        while not time_step.is_last():
            action_step = policy.action(time_step)
            time_step = environment.step(action_step.action)
            episode_return += time_step.reward
        total_return += episode_return
    avg_return = total_return / num_episodes
    return avg_return.numpy()[0]

# ---------- Entrenamiento principal ----------
for i in range(num_iterations):
    for _ in range(collect_steps_per_iteration):
        collect_step(train_env, collect_policy, replay_buffer)

    experience, _ = next(iterator)
    train_loss = agent.train(experience).loss

    if i % log_interval == 0:
        print(f"Iteración {i}, Pérdida: {train_loss.numpy():.4f}")

    if i % eval_interval == 0:
        avg_return = compute_avg_return(eval_env, agent.policy)
        print(f"→ Evaluación: retorno promedio = {avg_return:.2f}")

print("✅ Entrenamiento completado.")
````

A partir de la iteración 10000, el rendimiento empieza a recuperarse: primero alcanza un retorno de 44.2, y luego mejora significativamente hasta los 123.3 en la iteración 15000. Esto sugiere que, aunque el aprendizaje no fue completamente estable, el agente finalmente logró aprender una estrategia útil. Este tipo de oscilaciones es frecuente cuando se usan parámetros por defecto o no se ajustan con precisión los hiperparámetros clave (como el tamaño del replay buffer, la frecuencia de actualización del target network, o el número de pasos de actualización).

Es importante señalar que el replay_buffer tenía una capacidad bastante grande (100.000 transiciones), lo que puede hacer que se acumulen muchas experiencias antiguas que ya no reflejan el comportamiento actual del agente. Esto puede dificultar el aprendizaje si el entorno evoluciona rápido o si el agente cambia mucho su política con el tiempo. También puede influir el hecho de que solo se recolecta una transición por iteración, lo cual limita la velocidad con la que se actualizan las experiencias recientes.

En definitiva, aunque el entorno no llegó a considerarse "resuelto" oficialmente (lo cual requiere un retorno promedio ≥ 475 en 100 episodios consecutivos), los resultados indican que el agente sí aprendió una política efectiva, aunque de forma algo inestable. La red neuronal utilizada (dos capas densas de 100 y 50 unidades) y la configuración del agente parecen razonables para este entorno, y probablemente con un ajuste más fino se podría alcanzar un rendimiento mucho más alto y estable.

En resumen, el agente no fue perfecto, pero cumplió bien su función. Se entrenó con una configuración básica y, aun así, logró aprender a comportarse de manera bastante competente. Una optimización más cuidadosa de los hiperparámetros y el esquema de exploración probablemente permitiría mejorar estos resultados significativamente.

##1.Definici√≥n

El aprendizaje por refuerzo (RL) es una rama del aprendizaje autom√°tico en la que un agente aprende a tomar decisiones interactuando con su entorno. ##4.Episodio y trayectoria

##5.Estructura subyacente de un entorno

##6.Objetivo del agente

El objetivo principal del agente en RL es aprender una pol√≠tica √≥ptima que maximice el retorno o recompensa acumulada a lo largo del tiempo.
````python
import gymnasium as gym
from gymnasium.wrappers import TimeLimit

#creamos el entorno con l√≠mite de 500 pasos por episodio y render
base_env = gym.make('CliffWalking-v0')
env = TimeLimit(
    gym.make('CliffWalking-v0', render_mode='human'),
    max_episode_steps=500
)
# Entorno SIN render, solo para medir r√°pido
eval_env = TimeLimit(
    gym.make('CliffWalking-v0'),
    max_episode_steps=500
)
````

Con esto, podemos pasar a concretar cada una de las pol√≠ticas

###2.1 Pol√≠tica aleatoria

La pol√≠tica aleatoria no mira el estado. Simplemente pide una acci√≥n al espacio de acciones:

````python
def random_policy(env, state):
    """
    devuelve siempre una acci√≥n al azar
    del espacio de acciones.
    """
    return env.action_space.sample()
````

###2.2 Pol√≠tica √≥ptima manual

Como conocemos la cuadr√≠cula, escribimos una funci√≥n que, paso a paso, sigue la ruta segura (subir al borde, avanzar y bajar al final):

````python
def optimal_policy(env, state):
    """
    Sube a fila 2, avanza hasta columna 11 y baja a la meta.
    """
    ncols = 12
    row, col = divmod(state, ncols)

    # 1) En (3,0) subir
    if (row, col) == (3, 0):
        return 0          # arriba

    # 2) En fila 2 avanzar hasta col 11 (no 10)
    if row == 2 and col < 11:
        return 1          # derecha

    # 3) En (2,11) bajar a la meta
    if (row, col) == (2, 11):
        return 2          # abajo

    # 4) Cualquier otro caso (p. ej. ya en meta): mover derecha
    return 1
````

###2.3 Evaluaci√≥n sobre 100 episodios

Definimos una funci√≥n gen√©rica para correr una pol√≠tica durante N episodios y calcular la recompensa media:

````python
def evaluate_policy(env, policy_fn, episodes=100, render_one=False):
    rewards = []
    for ep in range(episodes):
        state, _ = env.reset()
        done = False
        total_r = 0

        while not done:
            action = policy_fn(env, state)
            state, r, terminated, truncated, _ = env.step(action)
            total_r += r
            done = terminated or truncated
            if render_one and ep == 0:
                env.render()

        rewards.append(total_r)

    avg = sum(rewards) / len(rewards)
    std = (sum((r-avg)**2 for r in rewards)/len(rewards))**0.5
    return avg, std
````

Y finalmente lo ejecutamos (tarda un ratillo)

````python
avg_rand, std_rand = evaluate_policy(eval_env, random_policy,
                                     episodes=100, render_one=False)
avg_opt,  std_opt  = evaluate_policy(eval_env, optimal_policy,
                                     episodes=100, render_one=False)

print(f"Aleatoria ‚Üí {avg_rand:.2f} ¬± {std_rand:.2f}")
print(f"√ìptima    ‚Üí {avg_opt:.2f}  ¬± {std_opt:.2f}")
````

Los n√∫meros hablan por s√≠ solos. Con la pol√≠tica aleatoria el agente acumula, de media, unos -4 800 puntos por episodio (desviaci√≥n ‚âà 1 400).

¬øPor qu√© tanto castigo? Durante los 500 pasos que permite el wrapper TimeLimit el agente deambula sin un criterio, pisa el precipicio en muchas ocasiones (-100 cada vez) y, aun cuando no cae, sigue recibiendo -1 por cada movimiento. El resultado t√≠pico es la suma de 500 penalizaciones peque√±as (-500) m√°s decenas de ca√≠das, de ah√≠ la media tan negativa y la gran variabilidad, en unos episodios se despe√±a m√°s que en otros.

La pol√≠tica ‚Äú√≥ptima‚Äù hard-coded, en cambio, finaliza cada partida con -13 puntos exactos y desviaci√≥n nula. Sube un paso, avanza junto al borde superior hasta la columna 11 y baja a la meta, son trece movimientos seguros, sin precipicio y sin truncamiento. Con ese simple conocimiento del entorno el rendimiento mejora tres √≥rdenes de magnitud respecto a la estrategia ciega.

En resumen, CliffWalking muestra c√≥mo la falta de planificaci√≥n combina una penalizaci√≥n constante (-1 por paso) con castigos puntuales muy altos (-100) y hunde el retorno; basta una ruta informada para eliminar casi todo el da√±o. El experimento ilustra la importancia, en aprendizaje por refuerzo, de aprender o incorporar r√°pidamente una pol√≠tica que evite transiciones de alto riesgo en entornos con recompensas tan asim√©tricas.

Los hiperpar√°metros los fijaremos de forma conservadora, Œ±=0,1 para que los valores se ajusten con suavidad, Œ≥=0,99 porque los episodios duran a lo sumo 500 pasos y conviene valorar casi por igual recompensa presente y futura, y un œµ que arranca en 1 y decae geom√©tricamente hasta 0,01 para ir pasando gradualmente de la exploraci√≥n a la explotaci√≥n.

Lo aplicamos sobre CliffWalking-v0, entorno utilizado en el apartado anterior, un escenario cl√°sico para medir la diferencia entre comportamientos on-policy y off-policy. El l√≠mite de 500 pasos impuesto con el wrapper TimeLimit convierte cada partida en un episodio acotado y aporta una referencia clara, incluso sin despe√±arse, una pol√≠tica que deambule pierde ya 500 puntos, as√≠ que cualquier mejora es f√°cil de cuantificar, y as√≠ tambi√©n podemos evaluarlo respecto a las pol√≠ticas anteriores.

````python
import random
import numpy as np

#hiperparametros
ALPHA      = 0.1
GAMMA      = 0.99
EPSILON    = 1.0
EPS_DECAY  = 0.9995
MIN_EPS    = 0.01

env = TimeLimit(gym.make("CliffWalking-v0"), max_episode_steps=500)
nS, nA = env.observation_space.n, env.action_space.n
Q = np.zeros((nS, nA))          #tabla Q(s,a)

def epsilon_greedy(state):
    if random.random() < EPSILON:
        return env.action_space.sample()
    return np.argmax(Q[state])

#entrenamiento
EPISODES = 5_000
for _ in range(EPISODES):
    s, _ = env.reset()
    done = False
    while not done:
        a  = epsilon_greedy(s)
        s2, r, term, trunc, _ = env.step(a)
        td_target = r + GAMMA * np.max(Q[s2])
        Q[s, a]  += ALPHA * (td_target - Q[s, a])
        s = s2
        done = term or trunc
        # update Œµ
        global EPSILON
        EPSILON = max(MIN_EPS, EPSILON * EPS_DECAY)

def run_policy(episodes=100):
    successes = 0
    returns   = []
    for _ in range(episodes):
        s,_ = env.reset()
        done = False
        G = 0
        while not done:
            a = np.argmax(Q[s])      # pol√≠tica greedy aprendida
            s, r, term, trunc, _ = env.step(a)
            G += r
            done = term or trunc
        returns.append(G)
        if term: successes += 1
    return successes, np.mean(returns)

succ, avg_ret = run_policy()
print(f"Resuelto {succ}/100 episodios ¬∑ recompensa media = {avg_ret:.1f}")
````

Tras 5 000 episodios de entrenamiento la tabla Q converge y, al ejecutar la pol√≠tica codiciosa resultante en 100 ensayos, el agente llega a la meta en los 100 episodios y obtiene siempre una recompensa de ‚Äì13. Ese valor coincide exactamente con el coste te√≥rico de la ruta segura (trece movimientos √ó -1) y la desviaci√≥n t√≠pica es nula porque la pol√≠tica aprendida es determinista y nunca pisa el precipicio.

El resultado obtenido es el mismo que en el algoritmo √≥ptimo del apartado anterior, un algoritmo dise√±ado expclusivamente para este problema teniendo en cuenta todos los factores.

El experimento nos muestra, en definitiva, c√≥mo un algoritmo cl√°sico y muy compacto como Q-Learning basta para resolver completamente un MDP discreto cuando el espacio de estados es manejable y la pol√≠tica dispone de la exploraci√≥n necesaria para descubrir la estrategia libre de riesgo, igualando el resultado d eun algoritmo dise√±ado √∫nicamente para este fin.

### 1. Algoritmos de Deep Reinforcement Learning

El aprendizaje por refuerzo profundo (Deep Reinforcement Learning, DRL) combina el marco del aprendizaje por refuerzo tradicional con redes neuronales profundas, lo que permite a los agentes aprender directamente a partir de entradas de alta dimensi√≥n, como im√°genes o estados complejos. A diferencia del aprendizaje supervisado, donde se necesita un conjunto de datos etiquetado, DRL aprende por ensayo y error, recibiendo recompensas o penalizaciones seg√∫n las acciones que ejecuta en un entorno.

Este paradigma se basa en el Proceso de Decisi√≥n de Markov (MDP), donde un agente observa un estado del entorno, ejecuta una acci√≥n, recibe una recompensa y transita a un nuevo estado. El objetivo es encontrar una pol√≠tica √≥ptima que maximice la recompensa acumulada a lo largo del tiempo, ponderada por un factor de descuento que da m√°s importancia a las recompensas inmediatas.

En DRL existen dos grandes familias de algoritmos:

- **Deep Q-Networks (DQN)**: Son m√©todos basados en valores. Usan una red neuronal para aproximar la funci√≥n Q(s, a), que estima la recompensa esperada de realizar una acci√≥n en un estado y seguir luego una pol√≠tica √≥ptima. Emplean t√©cnicas como experiencia repetida (replay buffer) y redes objetivo para estabilizar el aprendizaje. Ejemplos destacados son DQN, Double DQN (que reduce el sobreajuste), Dueling DQN (que separa el valor del estado y la ventaja de cada acci√≥n) y Prioritized Experience Replay, que mejora la eficiencia del muestreo. Son adecuados para entornos con espacios de acci√≥n discretos y definidos.

- **Policy Gradient**: Son m√©todos que optimizan directamente la pol√≠tica del agente, sin necesidad de estimar expl√≠citamente el valor de cada acci√≥n. En lugar de buscar la mejor acci√≥n desde la Q-function, la pol√≠tica se representa como una red neuronal que produce distribuciones de probabilidad sobre las acciones. Ejemplos cl√°sicos son REINFORCE, y versiones m√°s estables como Actor-Critic, A2C/A3C, PPO (Proximal Policy Optimization), que restringe los cambios bruscos en la pol√≠tica, o SAC (Soft Actor-Critic), que maximiza simult√°neamente la recompensa y la entrop√≠a para fomentar la exploraci√≥n. Estos m√©todos son m√°s apropiados para espacios de acci√≥n continuos o de gran tama√±o y permiten representar pol√≠ticas estoc√°sticas.

Mientras DQN aprende ‚Äúcu√°nto vale hacer algo‚Äù, Policy Gradient aprende directamente ‚Äúqu√© hacer‚Äù. Ambos enfoques tienen fundamentos comunes (como la ecuaci√≥n de Bellman), pero difieren en c√≥mo se modela y mejora la pol√≠tica del agente.

### 3. Descripci√≥n del algoritmo

El algoritmo Deep Q-Network (DQN) es una mejora del algoritmo cl√°sico Q-Learning, creado para enfrentar problemas donde el espacio de estados es muy grande o incluso continuo, lo que hace imposible usar una tabla para almacenar los valores de acci√≥n-estado Q(s,a). Para esto, DQN utiliza redes neuronales profundas que sirven como aproximadores de la funci√≥n Q(s,a).

El proceso comienza con un agente que interact√∫a con un entorno. En cada paso del proceso, el agente observa el estado actual s del entorno y debe elegir una acci√≥n a. Esta elecci√≥n no es arbitraria; el agente sigue una pol√≠tica de exploraci√≥n y explotaci√≥n llamada epsilon-greedy, que consiste en elegir la acci√≥n con el mejor valor estimado la mayor parte del tiempo (explotaci√≥n), pero tambi√©n de vez en cuando probar acciones aleatorias (exploraci√≥n) para descubrir nuevas estrategias. La probabilidad de exploraci√≥n est√° controlada por el par√°metro epsilon, que suele empezar alto y disminuir progresivamente.

Despu√©s de ejecutar la acci√≥n a, el entorno devuelve una recompensa r y un nuevo estado s'. Estos datos forman una experiencia completa (s, a, r, s'), que se guarda en una memoria llamada replay buffer o buffer de experiencia. Guardar estas experiencias permite que, m√°s adelante, el agente pueda entrenar la red neuronal usando datos de diferentes momentos y romper la correlaci√≥n temporal que existe si solo usara los datos en orden cronol√≥gico, lo cual mejora la estabilidad y eficiencia del aprendizaje.

La red neuronal profunda que se utiliza en DQN toma como entrada el estado s y devuelve los valores Q estimados para todas las acciones posibles en ese estado. Es decir, predice qu√© tan buena es cada acci√≥n en ese momento. Sin embargo, para evitar que el entrenamiento sea inestable o divergente, DQN utiliza dos redes neuronales con la misma arquitectura:

- La red principal Q(s,a; Œ∏) con par√°metros Œ∏, que es la que se entrena continuamente y genera las predicciones.

- La red objetivo Q(s,a; Œ∏‚àí) con par√°metros Œ∏‚àí, que es una copia de la red principal pero cuyos par√°metros se mantienen fijos durante un n√∫mero determinado de pasos. Cada cierto tiempo, los par√°metros Œ∏ de la red principal se copian a la red objetivo. Esto ayuda a que el c√°lculo de las metas para el entrenamiento sea m√°s estable.

El objetivo del entrenamiento es minimizar la diferencia entre las predicciones de la red principal y el valor esperado de la recompensa futura, definido por la ecuaci√≥n de Bellman adaptada a DQN:

Aqu√≠, gamma es un factor de descuento que valora cu√°nto se priorizan las recompensas futuras respecto a las inmediatas (un valor entre 0 y 1).

Para poder entrenar una red neuronal, necesitamos una funci√≥n de p√©rdida o coste (loss or cost function), la cual definimos como el cuadrado de la diferencia entre ambos lados de la ecuaci√≥n de Bellman:

image.png

El proceso de entrenamiento en cada iteraci√≥n incluye los siguientes pasos:

1. **Selecci√≥n de acci√≥n**: El agente elige una acci√≥n a en el estado actual s siguiendo la pol√≠tica epsilon-greedy basada en los valores que devuelve la red principal.

2. **Interacci√≥n con el entorno**: Ejecuta la acci√≥n a, recibe la recompensa r y observa el nuevo estado s'.

3. **Almacenamiento de experiencia**: La tupla (s, a, r, s') se guarda en el replay buffer.

4. **Muestreo**: Se selecciona aleatoriamente un conjunto peque√±o de experiencias (minibatch) del replay buffer para el entrenamiento.

5. **C√°lculo del valor objetivo**: Para cada experiencia del minibatch, se calcula el valor objetivo Q target usando la red objetivo.

6. **C√°lculo de la p√©rdida**: Se calcula la funci√≥n de p√©rdida como el error cuadr√°tico medio (MSE) entre los valores Q estimados por la red principal y los valores objetivo.

7. **Actualizaci√≥n de la red principal**: Se ajustan los pesos de la red principal para minimizar la funci√≥n de p√©rdida, utilizando t√©cnicas de retropropagaci√≥n y optimizadores como Adam o SGD.

8. **Actualizaci√≥n de la red objetivo**: Cada cierto n√∫mero de pasos, se copian los pesos de la red principal a la red objetivo para mantener la estabilidad del aprendizaje.

Adem√°s, el par√°metro epsilon se va reduciendo gradualmente para que, al principio, el agente explore muchas acciones diferentes, y a medida que aprende, se enfoque m√°s en explotar lo aprendido para maximizar las recompensas.

Gracias a este m√©todo, DQN puede aprender a resolver tareas complejas en las que el agente debe tomar decisiones en entornos con muchos estados posibles, como en el famoso problema CartPole. All√≠, el agente aprende a mover un carrito para mantener un palo equilibrado, mejorando su desempe√±o poco a poco hasta lograr mantener el palo estable durante largos periodos.

````python
import os
os.environ['TF_USE_LEGACY_KERAS'] = '1'

import tensorflow as tf
import numpy as np
from tf_agents.agents.dqn import dqn_agent
from tf_agents.networks import q_network
from tf_agents.utils import common
from tf_agents.environments import suite_gym, tf_py_environment
from tf_agents.replay_buffers import tf_uniform_replay_buffer
from tf_agents.policies import random_tf_policy
from tf_agents.trajectories import trajectory
from tf_agents.metrics import tf_metrics
from tf_agents.drivers import dynamic_step_driver
from tf_agents.policies import policy_saver

# ---------- Configuraci√≥n general ----------
env_name = "CartPole-v1"
num_iterations = 20000
initial_collect_steps = 1000
collect_steps_per_iteration = 1
replay_buffer_max_length = 100000

batch_size = 64
learning_rate = 1e-3
log_interval = 1000
eval_interval = 5000
fc_layer_params = (100, 50)

# ---------- Crear entornos ----------
train_env = tf_py_environment.TFPyEnvironment(suite_gym.load(env_name))
eval_env = tf_py_environment.TFPyEnvironment(suite_gym.load(env_name))

# ---------- Crear QNetwork ----------
q_net = q_network.QNetwork(
    input_tensor_spec=train_env.observation_spec(),
    action_spec=train_env.action_spec(),
    fc_layer_params=fc_layer_params
)

# ---------- Crear agente ----------
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
global_step = tf.compat.v1.train.get_or_create_global_step()

agent = dqn_agent.DqnAgent(
    time_step_spec=train_env.time_step_spec(),
    action_spec=train_env.action_spec(),
    q_network=q_net,
    optimizer=optimizer,
    epsilon_greedy=0.1,
    n_step_update=1,
    target_update_tau=0.05,
    target_update_period=200,
    td_errors_loss_fn=common.element_wise_squared_loss,
    gamma=0.99,
    train_step_counter=global_step
)
agent.initialize()

# ---------- Replay buffer ----------
replay_buffer = tf_uniform_replay_buffer.TFUniformReplayBuffer(
    data_spec=agent.collect_data_spec,
    batch_size=train_env.batch_size,
    max_length=replay_buffer_max_length
)

# ---------- Pol√≠ticas ----------
random_policy = random_tf_policy.RandomTFPolicy(train_env.time_step_spec(), train_env.action_spec())
eval_policy = agent.policy
collect_policy = agent.collect_policy

# ---------- Recolecci√≥n inicial ----------
def collect_step(environment, policy, buffer):
    time_step = environment.current_time_step()
    action_step = policy.action(time_step)
    next_time_step = environment.step(action_step.action)
    traj = trajectory.from_transition(time_step, action_step, next_time_step)
    buffer.add_batch(traj)

for _ in range(initial_collect_steps):
    collect_step(train_env, random_policy, replay_buffer)

# ---------- Dataset para entrenamiento ----------
dataset = replay_buffer.as_dataset(
    num_parallel_calls=3,
    sample_batch_size=batch_size,
    num_steps=2
).prefetch(3)
iterator = iter(dataset)

# ---------- Evaluaci√≥n ----------
def compute_avg_return(environment, policy, num_episodes=10):
    total_return = 0.0
    for _ in range(num_episodes):
        time_step = environment.reset()
        episode_return = 0.0
        while not time_step.is_last():
            action_step = policy.action(time_step)
            time_step = environment.step(action_step.action)
            episode_return += time_step.reward
        total_return += episode_return
    avg_return = total_return / num_episodes
    return avg_return.numpy()[0]

# ---------- Entrenamiento principal ----------
for i in range(num_iterations):
    for _ in range(collect_steps_per_iteration):
        collect_step(train_env, collect_policy, replay_buffer)

    experience, _ = next(iterator)
    train_loss = agent.train(experience).loss

    if i % log_interval == 0:
        print(f"Iteraci√≥n {i}, P√©rdida: {train_loss.numpy():.4f}")

    if i % eval_interval == 0:
        avg_return = compute_avg_return(eval_env, agent.policy)
        print(f"‚Üí Evaluaci√≥n: retorno promedio = {avg_return:.2f}")

print("‚úÖ Entrenamiento completado.")
````

## 5. Describir el uso de la biblioteca de aprendizaje para la resoluci√≥n del problema

Para implementar un agente DQN en el entorno CartPole-v1 utilizando TF-Agents, se dise√±√≥ una estructura modular que sigue las buenas pr√°cticas del aprendizaje por refuerzo, permitiendo separar claramente los componentes del entorno, la red neuronal, el agente y el proceso de entrenamiento.

Primero, se carga el entorno CartPole-v1 mediante suite_gym.load y se convierte con tf_py_environment.TFPyEnvironment para que funcione con tensores dentro del ecosistema de TensorFlow. Esto se hace tanto para el entorno de entrenamiento train_env como para el de evaluaci√≥n eval_env.

El agente se basa en una red neuronal profunda (q_net) implementada con la clase QNetwork, que internamente usa capas Dense de Keras. En este caso, la red consta de dos capas ocultas con 100 y 50 unidades respectivamente, ambas con activaci√≥n ReLU. La capa de salida tiene una neurona por acci√≥n disponible en el entorno, lo que permite estimar los valores Q para cada posible acci√≥n en un estado dado.

Como optimizador se emplea Adam con una tasa de aprendizaje (learning_rate) de 1e-3, valor que controla la magnitud de los pasos de actualizaci√≥n de los pesos. El agente se instancia con dqn_agent.DqnAgent, al que se le pasan varios argumentos clave:

- Las especificaciones de tiempo (time_step_spec) y de acci√≥n (action_spec) del entorno.

- La red Q definida anteriormente (q_network).

- Una pol√≠tica de exploraci√≥n Œµ-greedy con epsilon_greedy = 0.1, que garantiza cierto grado de aleatoriedad en las acciones durante el entrenamiento.

- Una funci√≥n de p√©rdida basada en el error cuadr√°tico medio (common.element_wise_squared_loss) entre el valor estimado y el valor objetivo.

- Par√°metros para la actualizaci√≥n del target network, como target_update_tau = 0.05 y target_update_period = 200, que ayudan a estabilizar el aprendizaje.

- Un contador global de pasos (train_step_counter) que registra el avance del entrenamiento.

Una vez creado el agente, se define un replay buffer con TFUniformReplayBuffer, el cual almacena las transiciones observadas para su posterior reutilizaci√≥n. Esta t√©cnica ayuda a romper la correlaci√≥n temporal entre las muestras de entrenamiento, lo que mejora la estabilidad del aprendizaje.

Inicialmente, se recolectan 1000 pasos usando una pol√≠tica completamente aleatoria (RandomTFPolicy) para poblar el buffer con experiencias variadas. Esto se hace mediante la funci√≥n collect_step, que genera trayectorias a partir de las interacciones entre el entorno y la pol√≠tica, y las almacena en el buffer.

Durante el entrenamiento, se construye un dataset a partir del buffer utilizando as_dataset, donde se especifican:

- sample_batch_size = 64: n√∫mero de muestras por lote.

- num_steps = 2: longitud de las trayectorias (estado, acci√≥n, siguiente estado).

- prefetch(3): para precargar datos y hacer m√°s eficiente el entrenamiento.

En cada iteraci√≥n del bucle principal, se recolecta una nueva transici√≥n (collect_steps_per_iteration = 1) usando la pol√≠tica actual del agente (collect_policy), se extrae un lote del buffer, y se entrena el agente con agent.train(experience). Adem√°s, cada cierto n√∫mero de pasos (log_interval), se imprime el valor de la p√©rdida para monitorizar el aprendizaje. Tambi√©n se eval√∫a el rendimiento del agente cada 5000 iteraciones usando su pol√≠tica aprendida, obteniendo un retorno promedio en varios episodios de prueba.

En cuanto a los hiperpar√°metros usados:

- learning_rate = 1e-3: velocidad de aprendizaje del optimizador.

- fc_layer_params = (100, 50): arquitectura de la red Q con dos capas ocultas.

- batch_size = 64: tama√±o del lote de entrenamiento.

- num_iterations = 20000: n√∫mero total de iteraciones de entrenamiento.

- replay_buffer_max_length = 100000: tama√±o m√°ximo del buffer de experiencia.

- collect_steps_per_iteration = 1: pasos recolectados por cada iteraci√≥n.

Gracias a esta implementaci√≥n, el agente es capaz de aprender una pol√≠tica efectiva para mantener el equilibrio del palo sobre el carro, utilizando aprendizaje por refuerzo profundo basado en DQN y la infraestructura de TF-Agents. Aunque no se us√≥ una configuraci√≥n hiperoptimizda, el sistema logra buenos resultados debido a una arquitectura adecuada y una integraci√≥n s√≥lida entre los m√≥dulos del entorno, la red, el agente y el entrenamiento.

### 6. Indicar c√≥mo se adapta el uso de la biblioteca

La implementaci√≥n con TF-Agents refleja fielmente la estructura te√≥rica de Deep Q-Learning. El agente DqnAgent encapsula los elementos clave del algoritmo: una red neuronal que estima la funci√≥n Q, una pol√≠tica basada en esta red para seleccionar acciones, y un mecanismo de actualizaci√≥n basado en la ecuaci√≥n de Bellman. La recolecci√≥n de experiencias (estado, acci√≥n, recompensa, siguiente estado) se realiza paso a paso con collect_step, en coherencia con el proceso de exploraci√≥n del entorno. Estas transiciones se almacenan en un buffer de repetici√≥n (TFUniformReplayBuffer), lo que permite aplicar aprendizaje por lotes desordenados, rompiendo la correlaci√≥n temporal entre muestras y estabilizando el entrenamiento. Adem√°s, la separaci√≥n entre la red actual y la red objetivo, aunque gestionada internamente por TF-Agents, sigue el principio te√≥rico de reducir la oscilaci√≥n de los valores Q durante la retropropagaci√≥n. Por √∫ltimo, el uso de una pol√≠tica aleatoria al inicio favorece la exploraci√≥n del espacio de estados, como se recomienda en la pr√°ctica del aprendizaje por refuerzo. As√≠, TF-Agents no solo facilita la implementaci√≥n, sino que permite aplicar de forma estructurada y coherente todos los elementos del algoritmo DQN descrito te√≥ricamente.
